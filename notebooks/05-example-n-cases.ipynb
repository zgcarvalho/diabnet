# Examples of individual risk prediction 

In this notebook we split the test set into four groups and get randomly 16 individuals of each one. The groups are:
 
* Elderly and positive (age > 60 and diabetic)
* Elderly and negative (age > 60 and not diabetic) 
* Young and positive (age < 40 and diabetic)
* Young and negative (age < 40 and not diabetic)

The age for positive patients corresponds to the age at the first positive diagnostic. So these patients have at least one negative exam at younger ages.

The goal of this notebook is observe characteristics of the risk prediction curve for individuals of each group.


## Elderly and positive

Individuals with first positive diagnostic after 60 years of age.

```python
fig = plot_elderly_and_positives(r)
if SAVE:
    for img_format in img_formats:
        fig.savefig(f"images/notebook5-plot-old-pos.{img_format}", bbox_inches="tight")
```

## Elderly and negative

Non-diabetic individual over the age of 60.

```python
fig = plot_elderly_and_negatives(r)
if SAVE:
    for img_format in img_formats:
        fig.savefig(f"images/notebook5-plot-old-neg.{img_format}", bbox_inches="tight")
```

## Young and positive

Individuals with first positive diagnostic before 40 years of age.

```python
fig = plot_young_and_positives(r)
if SAVE:
    for img_format in img_formats:
        fig.savefig(f"images/notebook5-plot-young-pos.{img_format}", bbox_inches="tight")
```

## Young and negative

Non-diabetic individual under the age of 40.

```python
fig = plot_young_and_negatives(r, compact=False)
if SAVE:
    for img_format in img_formats:
        fig.savefig(f"images/notebook5-plot-young-neg.{img_format}", bbox_inches="tight")
```
