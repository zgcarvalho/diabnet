```python
# load a trained model
net = load('../diabnet/models/model-layer01-FEAT-SELECT-50BW-000.pth')
```

```python
# selected features used to train the model above
c = ["snp_1126","snp_2444","snp_1","snp_1757","snp_9","snp_11","snp_2454","snp_10","snp_376","snp_8",
"snp_715","snp_1417","snp_1485","snp_4","snp_2","snp_3","snp_6","snp_5","snp_469","snp_1208",
"snp_1780","snp_790","snp_1301","snp_2208","snp_966","snp_992","snp_577","snp_1938","snp_1494","snp_2478",
"snp_1843","snp_2241","snp_1045","snp_368","snp_1652","snp_299","snp_615","snp_1150","snp_1479","snp_1461",
"snp_234","snp_1166","snp_7","snp_1283","snp_1773","snp_1272","snp_359","snp_954","snp_2371","snp_1509",
"snp_2474","snp_2472","snp_2470","snp_2469","snp_2462","snp_2459","snp_2458","snp_2450","snp_2449","snp_2442",
"snp_2439","snp_2434","snp_2432","snp_2431","snp_2430","snp_2428","snp_2427","snp_2420","snp_2419","snp_2417",
"snp_2412","snp_2405","snp_2402","snp_2399","snp_2395","snp_2384","snp_2377","snp_2375","snp_2369","snp_2368",
"snp_2362","snp_2358","snp_2357","snp_2354","snp_2352","snp_2349","snp_2348","snp_2337","snp_2326","snp_2323",
"snp_2318","snp_2317","snp_2316","snp_2315","snp_2309","snp_2305","snp_2302","snp_2292","snp_2282","snp_2275","AGE"]
```

```python
def encode_features(feat, feat_names, age=-1, bmi=-1):
    """
    Encode features into a matrix to be used as input to the net (neural network)
    """
    m = np.zeros((1,2, len(feat)))
    for i in range(len(feat_names)):
        if (feat_names[i] == "AGE"):
            if age == -1:
                m[0,0,i] = feat[i]
            else: 
                m[0,0,i] = age   
        elif (feat_names[i] == "BMI"):
            if bmi == -1:
                m[0,0,i] = feat[i]
            else: 
                m[0,0,i] = bmi 
        else:
            if feat[i] == 0:
                m[0,0,i] = 2
            elif feat[i] == 2:
                m[0,1,i] = 2
            else:
                m[0,0,i], m[0,1,i] = 1, 1
    return torch.tensor(m, dtype=torch.float32)
```

```python
def f(x):
    """
    Help function to apply the net 30 times to each from a batch of values (x). Used in shap.KernelExplainer
    """
    # Set a fixed AGE to keep its influence equal to all individuals
    age = 50
    return np.array([np.median([torch.sigmoid(net(encode_features(i,c,age))).detach().numpy()[0,0] for _ in range(30)]) for i in x])
```

```python
# get data to analyze
dftrain = pd.read_csv('../datasets/train.csv')
X_train = dftrain[c].values
Y_train = dftrain["T2D"].values
dftest = pd.read_csv('../datasets/test.csv')
X_test = dftest[c].values
Y_test = dftest["T2D"].values
```

```python
# summarize data to reduce complexity and computational cost
X_summary = shap.kmeans(X_train, 100, round_values=True)
```

```python
# shap explainer model-agnostic 
explainer = shap.KernelExplainer(f, X_summary, link="logit")
```

```python
# calculate shap values to test data
shap_values = explainer.shap_values(X_test, nsamples=100, l1_reg="aic");
```

```python
# plot shape values to indidual (idv) 
# idvs[0:168] -> positives idvs[168:702] -> negatives
idv = 1 
shap.force_plot(explainer.expected_value, shap_values[idv,:], dftest[c].iloc[idv], link="logit")
```
