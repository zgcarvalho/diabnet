```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append("../")
from diabnet.apply import Predictor
from diabnet.data import get_feature_names
```

```python
def get_feature_names(fn, BMI=False):
    import pandas as pd 
    import numpy as np
    col_names = pd.read_csv(fn).columns.values
    # features = np.delete(col_names, ['id','BMI','T2D'])
    if BMI:
        features = np.setdiff1d(col_names, np.array(['id','T2D']), assume_unique=True)
    else:
        features = np.setdiff1d(col_names, np.array(['id','BMI','T2D']), assume_unique=True)
    return list(features)
```

```python
c = get_feature_names("../datasets/visits_new_test_positivo_1000_random_0.csv", BMI=False)
```

```python
pred = Predictor('../diabnet/models/teste.pth', c, 'data/negatives_older60_from_positivo_1000_random_0.csv')
```

```python
# 1286 positive
# 1292 negative
df = pd.read_csv("../datasets/visits_new_test_positivo_1000_random_0.csv")
id = 1149
features = df[c].iloc[id].values
label = df["T2D"].iloc[id]
```

```python
info = ["AGE", "T2D"]
df[info].iloc[id]
```

```python
probs = pred.patient(features, age=40)
neg = pred.negatives(age=40)
```

```python
sns.boxplot(x=['patient', 'negatives\n(age >= 60)'], y= [probs[:100], neg[:100]])
plt.ylim(0,1);
```

```python
neg_life = pred.negatives_life(samples=100)
```

```python
probs_life = pred.patient_life(features, samples=5000)
```

```python
plt.figure(figsize=(16,8))
plt.subplot(121)
sns.boxplot(x=[i for i in probs_life[1]], y=probs_life[0]);
plt.xlabel("patient")
plt.ylim(0,1);
plt.subplot(122)
sns.boxplot(x=[i for i in neg_life[1]], y=neg_life[0], showfliers=False);
plt.xlabel("negatives\n(age >= 60)")
plt.ylim(0,1);
plt.savefig("negative-lifetime-boxplots.png")
```
