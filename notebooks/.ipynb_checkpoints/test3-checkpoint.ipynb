```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import numpy as np
from sklearn.metrics import roc_auc_score, roc_curve, precision_recall_curve,f1_score, accuracy_score, balanced_accuracy_score, brier_score_loss
sys.path.append("../")
from diabnet.apply import Predictor
```

```python
def get_feature_names(fn, BMI=False):
    import pandas as pd 
    import numpy as np
    col_names = pd.read_csv(fn).columns.values
    # features = np.delete(col_names, ['id','BMI','T2D'])
    if BMI:
        features = np.setdiff1d(col_names, np.array(['id','T2D']), assume_unique=True)
    else:
        features = np.setdiff1d(col_names, np.array(['id','BMI','T2D']), assume_unique=True)
    return list(features)
```

```python
c = get_feature_names("../datasets/visits_checked_unique_test_positivo_1000_random_0.csv", BMI=False)
```

```python
pred = Predictor('../diabnet/models/teste-checked.pth', c, 'data/negatives_older60_from_positivo_1000_random_0.csv')
```

# Testes para o modelo

## Testes utilizando todo o conjunto de teste 

Os resultados abaixo utilizam todos os 488 indivíduos do conjunto de teste. 

No conjunto há 128 positivos para diabetes.

```python
# o conjunto de testes possui ids repetidos devido as visitas (idades diferentes) 
# abaixo é feita a leitura dos dados retirando a repetição de ids, mantendo a maior idade e estado diabetico caso tenha sido observado em alguma visita  
df = pd.read_csv("../datasets/visits_checked_unique_test_positivo_1000_random_0.csv").groupby('id').max()
features = df[c].values
labels = df['T2D'].values
```

```python
# as predições são feitas em diferentes idades para avaliar como isso influencia no resultado
# por default da funcao pred.patient cada predicao é realizada 100 vezes. O valor final do individuo é a mediana dessas prediçoes
predictions = [[np.median(pred.patient(f, age=age)) for f in features] for age in range(20,80,5)]
```

```python
aucs = [roc_auc_score(labels,pr) for pr in predictions]
for i in zip(range(20,80,5), aucs):
    print("Predictions using patient age = {} have AUC = {:.4f}".format(i[0],i[1]))
```

```python
f1s = [f1_score(labels,np.round(pr)) for pr in predictions]
for i in zip(range(20,80,5), f1s):
    print("Predictions using patient age = {} have F1-Score = {:.4f}".format(i[0],i[1]))
```

```python
accs = [accuracy_score(labels,np.round(pr)) for pr in predictions]
for i in zip(range(20,80,5), accs):
    print("Predictions using patient age = {} have Accuracy = {:.4f}".format(i[0],i[1]))
```

```python
baccs = [balanced_accuracy_score(labels,np.round(pr)) for pr in predictions]
for i in zip(range(20,80,5), baccs):
    print("Predictions using patient age = {} have Balanced Accuracy = {:.4f}".format(i[0],i[1]))
```

```python
for pr in predictions:
    fpr, tpr, threshold = roc_curve(labels,pr)
    plt.plot(fpr, tpr)
plt.ylim(0,1);
plt.xlim(0,1);
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# plt.savefig('test-auc-all-testset.png')
```

```python
for pr in predictions:
    precision, recall, _ = precision_recall_curve(labels, pr)
    plt.step(recall, precision,where='post')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.0])
plt.xlim([0.0, 1.0])
# plt.savefig('test-precision-recall-all-testset.png')
```

```python
precision, recall, threshold = precision_recall_curve(labels, predictions[7])
print(precision[338], recall[338], threshold[338])
```
