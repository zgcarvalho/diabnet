```python
import sys
sys.path.append("../")
from diabnet.model import load
import pandas as pd
import numpy as np
import torch
import matplotlib.pyplot as plt
import seaborn as sns
import shap
```

```python
def get_feature_names(fn, BMI=False):
    import pandas as pd 
    import numpy as np
    col_names = pd.read_csv(fn).columns.values
    # features = np.delete(col_names, ['id','BMI','T2D'])
    if BMI:
        features = np.setdiff1d(col_names, np.array(['id','T2D']), assume_unique=True)
    else:
        features = np.setdiff1d(col_names, np.array(['id','BMI','T2D']), assume_unique=True)
    return list(features)
```

```python
c = get_feature_names("../datasets/positivo_500_negative_500.csv", BMI=False)
```
