```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_orig = pd.read_csv("../data/datasets/visits_sp_unique_train_positivo_1000_random_0.csv", index_col=0)
```

```python
df_br = pd.read_csv("../data/datasets/maira/dnabr_hg19_t2d.csv", index_col=0)
```

```python
df_imp = pd.read_csv("./feat_importance_trainset_X50_lc7-2021-10-05.csv", index_col=0)
```

```python
c = [c for c in df_orig.columns if 'snp'in c]
```

```python
c_br = [c for c in df_br.columns if 'snp'in c]
```

```python
freq_orig = df_orig[c].sum()/(df_orig.count()[0]*2)
```

```python
freq_br = df_br[c_br].sum()/(df_br.count()[0]*2)
```

```python
plt.hist()
```

```python
plt.hist()
```

```python
importance = df_imp['X50_mean'].transpose()
```

```python
fig = plt.figure(dpi=300)
plt.scatter(importance[c], freq_orig[c])
plt.ylim(0,1)
plt.ylabel("SNP frequency")
fig.savefig("images/freq_original.png")
```

```python
fig = plt.figure(dpi=300)
plt.scatter(importance[c_br], freq_br[c_br])
plt.ylabel("SNP frequency")
fig.savefig("images/freq_dnabr.png")
```

```python
fig = plt.figure(dpi=300, figsize=(5,5))
plt.scatter(freq_br[c], freq_orig[c])
plt.xlim(-0.02,1.02)
plt.ylim(-0.02,1.02)
plt.xlabel("our dataset (snp freq)")
plt.ylabel("training dataset (snp freq)")
plt.plot([0,1], [-0,1], c='k', ls=':')
plt.hlines(0.5, -1, 2, color='r')
fig.savefig("images/freq_diff.png")
```
