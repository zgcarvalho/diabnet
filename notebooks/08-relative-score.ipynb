# Relative Score

A relative score is a measure of how a person's score compares to that of the population. However, the samples were taken from an ethnic group with one of the highest prevalences and, therefore, a single relative score would have a distortion due to distribution bias.   

Alternatively, we chose to calculate a relative score for each group, diabetic and non-diabetic.

Diabetic group relative score lies in the range of \[0,1] and represents the ratio of the diabetic group with predicted mean value (absolute score) BELOW the predicted value. All predictions are made at the same age.

Non-diabetic relative score is in the range of \[-1,0] and represents the ratio of the non-diabetic group with predicted mean value ABOVE the patient predicted value. As the other score, all preditions are made at the same age. However, unlike the diabetic group relative score where all positives are used, in calculating non-diabetics relative score, negative individuals younger than the target age are removed.
